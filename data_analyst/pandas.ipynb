```python
# 결측치 수와 비율 계산
missing_values = df.isna().sum()  # 결측치 수
missing_ratio = (df.isna().sum() / len(df)) * 100  # 결측치 비율

# 결과를 데이터프레임으로 만들어서 출력
missing_df = pd.DataFrame({
    '결측치 수': missing_values,
    '결측치 비율(%)': missing_ratio
})
print('\n결측치 수와 비율:')
missing_df
```

## 결측치 대응
---

결측치가 존재하는 Age, Cabin, Embarked에 대해 대응

### 결측치대응 1: Age

```python
df.Age.mean()
```

```python
df["Age"] = df.Age.fillna(df.Age.mean())
```

```python
df.Age.isna().sum()
```

### 결측치대응 2: Cabin

```python
df.Cabin.isna().sum()
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df.columns
```

### 결측치대응3: Embarked         

```python
df = df[~df.Embarked.isna()]
```

```python
df
```

```python
df.isna().sum().sum()
```
