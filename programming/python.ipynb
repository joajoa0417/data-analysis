```python
def make_filter(min_score):
    def filter_func(score):
        return score >= min_score
    return filter_func
```

```python
high_pass_filter = make_filter(80)
low_pass_filter = make_filter(50)
```

```python
scores = [85, 75, 60, 59, 95, 87]
```

```python
print("80점 이상:", list(filter(high_pass_filter, scores)))
print("50점 이상:", list(filter(low_pass_filter, scores)))
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("새로운 기능 1")
        func(*args, **kwargs)
        print("새로운 기능 2")
    return wrapper

def print_operation(x, y):
    print(x * 2 + y)
```

```python
decorated = my_decorator(print_operation)
```

```python
print("--- print_operation 함수 호출 ---")
print_operation(5, 7)

print("--- decorated 함수 호출 ---")
decorated(5, 7)
```

```python
class CountUp:
    def __init__(self, max):
        self.max = max
        self.current = 1

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        value = self.current * 3
        self.current += 1
        return value
```

```python
counter = CountUp(3)
for n in counter:
    print(n)
```
