```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.bok.or.kr/portal/singl/baseRate/list.do?dataSeCd=01&menuNo=200643")
driver.implicitly_wait(2)

path = '//*[@id="main-container"]/div[2]/div[2]/table/tbody/tr'

rows = driver.find_elements(By.XPATH, path)

results = []

for tr in rows:
    cells = tr.find_elements(By.TAG_NAME, 'td')
    row_data = [cell.text.strip() for cell in cells]
    if row_data:
        results.append(row_data)

driver.quit()

df = pd.DataFrame(results, columns=['연도', '일자', '기준금리'])

```

```python
from bs4 import BeautifulSoup
import requests
from itertools import zip_longest

url = 'https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=ai'
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

titles = soup.select('span.sds-comps-text.sds-comps-text-ellipsis-1.sds-comps-text-type-headline1')
brands = soup.select('span.sds-comps-text.sds-comps-text-type-body2.sds-comps-text-weight-sm')

news = []

for title, brand in zip_longest(titles, brands, fillvalue=None):
    news_title = title.get_text(strip=True) if title else '제목 없음'
    news_brand = brand.get_text(strip=True) if brand else '뉴스사 없음'
    news.append({'제목': news_title, '뉴스사': news_brand})

news_df = pd.DataFrame(news, columns=['제목', '뉴스사'])
news_df


    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()
driver.maximize_window()
driver.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query=ai')
driver.implicitly_wait(2)

# 기사 제목 a 태그는 대부분 news_tit 클래스를 가짐
titles = driver.find_elements(By.CSS_SELECTOR, 'a.news_tit')

news = []

for title in titles:
    news_title = title.text.strip()  # 텍스트 내용 추출
    news.append({'기사 제목': news_title})

driver.quit()

# DataFrame 생성
df = pd.DataFrame(news)
print(df)
```
