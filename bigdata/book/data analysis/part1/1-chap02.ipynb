## **Pandas**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "메뉴" : ['아메리카노', '카페라떼', '에스프레소', '카페모카', '바닐라라떼'],
    "가격" : [4500, 5000, 4000, 5900, 5300],
    "칼로리" : [10, 110, np.NaN, 210, np.NaN],
    "원두" : ['과테말라', '브라질', '과테말라', np.NaN, np.NaN]
})

df
```

```python
# Q1
kal_min = df['칼로리'].min()
df['칼로리'] = df['칼로리'].fillna(kal_min)
df
```

```python
# Q2
wondu_mode = df['원두'].mode()[0]
df['원두'] = df['원두'].fillna(wondu_mode)
df
```

```python
# Q3
cond = df['가격'] >= 5000
df[cond].shape[0]
```

```python
# Q4
df['이벤트가'] = df['가격'] * 0.5
df
```

```python
# Q5
df = df.drop('칼로리', axis=1)
df
```
