### **NewYork airBnB Data**

##### Q44. 데이터를 로드하고 상위 5개 컬럼을 출력하라.

```python
url = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df4 = pd.read_csv(url)
df4.head()
```

##### Q45. 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라.

```python
df4.groupby('host_name').size().head()
```

```python
df4['host_name'].value_counts().sort_index().head()
```

`sort_index()`와 `sort_values()`의 차이


* `sort_index()`는 사전 순으로 정렬
* `sort_values()`는 데이터프레임에서 해당 컬럼별 값의 개수를 기준으로 정렬

##### Q46. 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라.

```python
pd.DataFrame(df4['host_name'].value_counts())
```

```python
df4.groupby('host_name').size().\
    sort_values(ascending=False).to_frame().rename(columns={0:'counts'})
```

##### Q47. neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라.

```python
df4.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
```

`as_index()`는 grouping한 기준 컬럼들이 결과 데이터프레임 index로 설정

##### Q48. neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라.

```python
df4.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()\
.groupby('neighbourhood_group', as_index=False).max()
```

##### Q49. neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라.

```python
df4.groupby('neighbourhood_group')[['price']].agg(['mean','var','max','min'])
```

##### Q50. neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라.

```python
df4.groupby('neighbourhood_group')[['reviews_per_month']].agg(['mean','var','max','min'])
```

##### Q51. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라.

```python
df4.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean()
```

##### Q52. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라.

```python
df4.groupby(['neighbourhood','neighbourhood_group'])['price'].mean().unstack()
```

##### Q53. neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라.

```python
df4.groupby(['neighbourhood','neighbourhood_group'])['price'].mean().unstack().fillna(-999)
```

##### Q54. 데이터 중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라.

```python
df4[df4['neighbourhood_group'] == 'Queens'].groupby('neighbourhood')['price'].agg(['mean','var','max','min'])
```

##### Q55. 데이터 중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라.

```python
room = df4.groupby('neighbourhood_group')['room_type'].value_counts().unstack()
room.div(room.sum(axis=1), axis=0)
```
